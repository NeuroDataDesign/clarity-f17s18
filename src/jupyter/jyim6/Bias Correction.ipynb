```python
%matplotlib notebook
import matplotlib.pyplot as plt
import os
import numpy as np
import glob
import tifffile as tif
import SimpleITK as sitk
```

```python
CORRECTED_FILES_DIR = './ssd2/s3617_corrected_tiles/CH01flat/'
RAW_FILES_DIR = './zd/s3617_raw_tiles/'

CORRECTED_FILES_LOCS = sorted(glob.glob(CORRECTED_FILES_DIR+'*'))
RAW_FILES_CH = glob.glob(RAW_FILES_DIR+'*')
```

# Choosing a location and looking at the data

```python
corrected_loc_sample = CORRECTED_FILES_LOCS[0]
print "Corrected sample location: {}".format(corrected_loc_sample)
corrected_files_sample = sorted([f for f in glob.glob(corrected_loc_sample+'/*.tif')])
corrected_combined_tiff = sitk.ReadImage(corrected_files_sample)
corrected_combined_tiff_np = sitk.GetArrayFromImage(corrected_combined_tiff)
```

```python
raw_files_ch_sample = RAW_FILES_CH[0]
raw_files_loc = sorted([f for f in glob.glob(raw_files_ch_sample+'/LOC*')])[0]
print "Raw sample location: {}".format(raw_files_loc)
raw_files_sample = sorted([f for f in glob.glob(raw_files_loc+'/*.tif')])
raw_combined_tiff = sitk.ReadImage(raw_files_sample)
raw_combined_tiff_np = sitk.GetArrayFromImage(raw_combined_tiff)
```

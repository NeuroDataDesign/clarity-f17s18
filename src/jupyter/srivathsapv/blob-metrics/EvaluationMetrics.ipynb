# Evaluation metrics against 10 manually annotated subvolumes for HDoG

### Initialization code

```python
import csv
from BlobMetrics import BlobMetrics
import ipyvolume as ipv
import numpy as np
import sys
```

```python
def get_coords_from_csv(filepath):
  with open(filepath) as infile:
    rows = csv.reader(infile)
    coords = [[float(r[0]), float(r[1]), float(r[2])] for r in list(rows)]

  return coords
```

```python
def output_metrics_for_experiment(ground_truth_path, predictions_path, euclidean_distance_threshold=12):
    ground_truth_coords = get_coords_from_csv(ground_truth_path)
    predicted_coords = get_coords_from_csv(predictions_path)
    
    metrics = BlobMetrics(ground_truth_coords, predicted_coords, euclidean_distance_threshold)
    
    accuracy = metrics.accuracy()
    print('Accuracy: {}'.format(accuracy))
    
    precision = metrics.precision()
    print('Precision: {}'.format(precision))
    
    recall = metrics.recall()
    print('Recall: {}'.format(recall))
    
    fmeasure = metrics.f_measure()
    print('F-Measure: {}'.format(fmeasure))
    
    mse = metrics.mean_square_error()
    print('Mean Square Error: {}'.format(mse))
    
    metrics.plot_predictions_with_ground_truth()
    metrics.plot_predictions_per_ground_truth()
    metrics.plot_ground_truths_per_prediction()
```

```python
output_metrics_for_experiment('data/annotations/cell_detection_0.csv', 'data/predictions/cell_detection_0.csv')
```

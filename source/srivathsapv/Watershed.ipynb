## Watersheld algorithm

**Import image and display it**

```python
import numpy as np
import pylab
import mahotas as mh
cell = mh.imread('input.jpg')
pylab.imshow(cell)
pylab.show()
```

**Threshold the image with otsu binarization**

```python
T = mh.thresholding.otsu(cell)
out_image = (cell > T)
out_image = out_image.astype(np.uint8) * 255

labeled, nr_objects = mh.label(out_image > T)
print nr_objects
```

**Show the output image normalized to grayscale (multiply by 255)**

```python
pylab.imshow(labeled.astype(np.uint8) * 255)
pylab.jet()
pylab.show()
```
